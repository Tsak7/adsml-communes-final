## Task D: Probabilities

```python
high_elev = df[df["Elevation"] > 2000]
```

```python
valais_high = high_elev[high_elev["Canton"] == "VS"]  # sometimes "Valais" depending on dataset
```

```python
df["Canton"].unique()
```

```python
prob = len(valais_high) / len(high_elev)
prob
```

## Task D Conclusion

In this task, the objective was to compute a conditional probability related to elevation and canton membership. Specifically, we estimated the probability that a randomly selected commune with an elevation above 2000 meters belongs to the canton of Valais.

To do this, I first filtered the dataset to retain only communes with an elevation greater than 2000 meters. This step was necessary to restrict the sample space to high-altitude communes, as required by the conditional probability definition.

Next, I further filtered this subset to identify communes belonging to the canton of Valais. This allowed me to isolate the favorable outcomes within the previously defined condition.

Finally, I computed the probability as the ratio between the number of Valais communes in the high-elevation group and the total number of high-elevation communes.

This approach directly reflects the definition of conditional probability:
P(A | B) = P(A ∩ B) / P(B), where A is “commune in Valais” and B is “elevation > 2000m”.

The result highlights the strong relationship between high-altitude geography and the canton of Valais, which is consistent with its location in the Swiss Alps.

The probability represents the proportion of communes with elevation above 2000 meters that are located in the canton of Valais. This result highlights the strong association between high-altitude regions and Valais, which is consistent with its geographic position in the Swiss Alps.

## Task E Conclusion

In this task, the goal was to transform the categorical variable “Canton” into a numerical matrix representation that can be used in further data analysis or machine learning models.

To achieve this, I used the `pd.get_dummies()` function from pandas. This function automatically performs one-hot encoding, converting each unique canton into a separate binary column. Each row (commune) is assigned a value of 1 in the column corresponding to its canton and 0 in all others.

This approach was chosen because it is efficient, avoids manual looping, and ensures correctness when dealing with categorical variables. It also guarantees that the resulting matrix is compatible with mathematical operations commonly used in machine learning.

The resulting matrix allows each commune to be represented as a structured numerical vector, making it suitable for further applications such as clustering, regression, or classification tasks.
